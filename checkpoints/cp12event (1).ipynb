```python
#Output thetas of each pair as well as the associated momentum value

passed1 = []        #store eventIDs that have been searched already so iterate through each eventID only once.
passed2 = []
B = 0.5             #magnetic field strength given
L2 = 8              #total length
L1 = 4              
calc_p = []
true_p = []
event_vals = []

for i in tracker1Data['EventID']:
    #only iterate through unique events, store eventIDs that have been iterated through in a list and skip them in future loops
    if i in passed1:  
        continue 

    #iterate through all permutations of rows for a specific event.
    for j in tracker1Data[tracker1Data['EventID']==i].values:
        for k in tracker2Data[tracker2Data['EventID']==i].values:
            if ( math.fabs( j[2] - k[2] ) < 0.0001 ):    #compare thetas of all permutations of hits in event to find pairs
                print( "Pair found!" )
                deltaPhi = math.fabs( j[1] - k[1] )      #Calculate Phi
                
                s = math.sin(deltaPhi) * L1
              
                R = ((L2)**2)/(8*s)
               
                pt = 0.3 * B*R
                
               
                p = pt/math.sin(j[2])
                print('Theta angles of pair: ',j[2],k[2] )   
                print('Momentum is: ',p)
                
                #iterate through true events
                for l in truthData[truthData['EventID']==i].values:
                    
                    if (math.fabs(k[2] -  l[2]) < 0.0001 ): #compare thetas of tracker2 and true data
                        calc_p.append(p)          #store calculated momentum
                        true_p.append(l[3])       #store true momentum
                        event_vals.append(i)      #store corresponding event id
                        #print(p)
                        #print(l[3])
                
    

    
    passed1.append(i)
```

```python
#Calculate distribution
calc_p = np.array(calc_p)
true_p = np.array(true_p)
distr = (calc_p - true_p)/true_p
```

```python
plt.hist(distr, bins =800)
plt.xlim(-0.1,0.1)
plt.ylabel('Number of reconstructed particles')
plt.xlabel('(Calculated momentum - true momentum)/true momentum')

plt.show()
```

```python

#Calculate width
print('The standard deviation on this is {:.2f}'.format(np.std(distr)))

```

```python
df = pd.DataFrame({'event':event_vals,'true_p': true_p,'recon_p': calc_p})
```

```python
counts = df['event'].value_counts()       #count how many counts in each event
target= counts[counts==2].index        #define criteria for keeping event which is having two counts. 
df =df[df['event'].isin(target)]         #Discard events without 2 events
```

```python
results = df.groupby('event')['recon_p'].sum().reset_index()    #sum pairs of reconstructed momentums from same event 
```

```python
masses = np.array(results['recon_p'])

h,bins = np.histogram(masses,bins = 380)
plt.hist(masses,bins = 380)
plt.xlim(80,95)
plt.xlabel('Total reconstructed energy(GeV)')
plt.ylabel('Number of events')
plt.show()
```

```python
print('The mass of the particle is {:.2f}'.format(np.mean(masses)),'GeV')

#decay width here is the standard deviation
print('The standard deviation is {:.2f}'.format(np.std(masses)),'GeV')

```

Constrain to within 90 and 92 where the majority of points lie

```python
constrained_masses = masses[(masses>90)&(masses<92)]
print('The mass of the particle is {:.2f}'.format(np.mean(constrained_masses)),'GeV')
print('The standard deviation  is {:.2f}'.format(np.std(constrained_masses)),'GeV')
```

This is a z boson since the mass is close to the mass of the Z boson listed on PDG as 91.1876 GeV. This is greater than the calculated mean for our data.  The total width of the boson on the PDG is 2.4952 GeV and the partial width for the muon decays is 0.033662 *2.4952 = 0.08 GeV. This partial decay is significantly less than the width we found for our data above. These differences are expected because we made approximations in our data analysis, for example where we may have double counted particle pairs.
